```python
# 각 수강생에게 이메일 발송
for index, row in df.iterrows():
    name = row['이름']       # 수강생 이름이 들어있는 열
    email = row['이메일']    # 수강생 이메일이 들어있는 열
    
    # 메일 발송 함수 호출
    send_mail(s, ID, email, name, row)

# 발송 후 SMTP 연결 종료
s.quit()
print("모든 이메일 발송 완료")
```
